```python
#12 Movies per year
data.groupby('Year')['Movie_Title'].count()
```

```python
#13  Movies with rating > 8 & revenue > 100M
data[(data['Rating']>8) & (data['Revenue (Millions)']>100)]
```

```python
#14  Group by Director: avg Rating, Revenue
data.groupby('Director')[['Rating','Revenue (Millions)']].mean()
```

```python
#15 Group by Year: count, avg Rating
data.groupby('Year').agg(movie_count=('Rating','count'),avg_rating=('Rating','mean'))
```

```python
#16  Create a new column Profitability: a) Revenue divided by Duration b) Find top 10 most profitable movies
data['Profitability'] = data['Revenue (Millions)']/data['Runtime (Minutes)']
data
```

```python
data.nlargest(10,'Profitability')[['Movie_Title','Profitability']]
```

```python
#17  Create a column that categorizes movies by rating: a) Rating < 6: Poor b) 6 ≤ Rating < 7.5: Average c) Rating ≥ 7.5: Excellent
data['Movies_by_rating']=data['Rating'].apply(
     lambda x : "Poor" if x<6
    else "Average" if 6<=x<7.5
    else "Excellent"
)
data
```

```python
#18 Create a custom function to score movies based on: i) Apply row['Rating'] * 10 + row['Revenue (Millions)'] as High score Apply a function to dataset to assign the score
def Movie_Score(row):
  return row['Rating']*10 + row['Revenue (Millions)']
data['Score']=data.apply(Movie_Score,axis=1)
data
```

```python
data.sort_values('Score',ascending=False)
```
